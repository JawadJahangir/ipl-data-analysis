```python
import matplotlib.pyplot as plt
import pandas as pd

data={
    "Player": ["Kohli","Rohit","Dhoni","Warner","ABD"],
    "Runs": [6700,6000,5000,6200,5500],
    "Matches": [223,227,234,170,184]
}

df=pd.DataFrame(data)
df.to_csv("Dataset.csv")
```

```python
#Bar Chart (Runs by Player)
plt.bar(data["Player"],data["Runs"],color="skyblue")
plt.xlabel("Players",fontsize=15,labelpad=20,color="blue")
plt.ylabel("Runs",fontsize=15,labelpad=20,color="blue")
plt.title("Runs By Players",fontsize=20,color="red")
plt.show()
```

```python
#Horizontal Bar (Matches Played)
plt.barh(data["Player"],data["Matches"],color="green")
plt.title("Matches Played",fontsize=20,color="red")
plt.xlabel("Matches",fontsize=15,labelpad=20,color="red")
plt.ylabel("Players",fontsize=15,labelpad=20,color="red")
plt.show()
```

```python
#Line Plot (Runs Trend)
plt.plot(data["Player"],data["Runs"],color="orange",linewidth=2,marker="o",linestyle="dashed")
plt.title("Runs Trend",fontsize=20,color="red",fontweight="bold")
plt.xlabel("Players",fontsize=20,fontweight="bold")
plt.ylabel("Runs",fontsize=20,fontweight="bold")
plt.show()
```

```python
#Pie Chart (Run Contribution)
color=["red","green","yellow","orange","blue"]
explode=[0.1,0,0,0,0]
plt.pie(data["Runs"],colors=color,labels=data["Player"],
        explode=explode,
        autopct="%1.1f%%",
        textprops={"fontsize":15,"color":"pink"},
        startangle=90,
        shadow=True,
        wedgeprops={"linewidth":1,"edgecolor":"black"}
        )
plt.show()
```
